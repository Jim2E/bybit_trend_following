```python
def get_perp_dash():
    pair_names = np.sort(['AVAXUSDT','BTCUSDT','DYDXUSDT','FILUSDT','FTMUSDT','FXSUSDT','GMXUSDT','IMXUSDT','LOOKSUSDT','LTCUSDT','MINAUSDT','OPUSDT','RNDRUSDT','SOLUSDT'])

    perps_dash={'Perp': [],
                'Price': [],
                'Key Level': [],
                'Perc. to KL': [],
                'Sim USDT': [],
                'Wins': [],
                'Losses': [],
                'SL Rate':[],
                'TP Rate':[],
                'min_dt_local_top':[],
                'PNL Avg.':[],
                'PNL Std.':[]}

    for pair_name in pair_names:
        recv = tf_dynamic_run_w_grid_search(token=pair_name,rewind_time=30)
        for k,v in perps_dash.items():
            perps_dash[k].append(recv[k])
        print(f'Perp:{pair_name} \t USDT:{perps_dash["Sim USDT"][-1]} \t SL:{perps_dash["SL Rate"][-1]} \t TP:{perps_dash["TP Rate"][-1]} \t Gap(%):{100*perps_dash["Perc. to KL"][-1]}')

    perps_dash = pd.DataFrame(perps_dash)
    return perps_dash
```

```python
dash = get_perp_dash()
code_completed()
```

```python
dash.sort_values(by='Sim USDT',ascending=False).reset_index(drop=True)
```
