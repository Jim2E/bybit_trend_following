```python
def get_perp_dash():
    pair_names = np.sort(['AGIXUSDT','AVAXUSDT','BLURUSDT','DYDXUSDT','FETUSDT','FILUSDT','FTMUSDT','FXSUSDT','GMXUSDT','IMXUSDT','LOOKSUSDT','LTCUSDT','MINAUSDT','OPUSDT','RNDRUSDT','SOLUSDT'])

    perps_dash={'token': [],
                'last_date': [],
                'price': [],
                'last_rlvl': [],
                'perc_to_rlvl': [],
                'sim usdt': [],
                'sim wins': [],
                'sim losses': [],
                'sl rate':[],
                'tp rate':[],
                'dt_local_top':[]}

    for pair_name in pair_names:
        recv = tf_dynamic_run_w_grid_search(token=pair_name,rewind_time=30)
        for k,v in perps_dash.items():
            perps_dash[k].append(recv[k])
        print(f'Perp:{pair_name} \t USDT:{perps_dash["sim usdt"][-1]} \t SL:{perps_dash["sl rate"][-1]} \t TP:{perps_dash["tp rate"][-1]} \t Gap(%):{100*perps_dash["perc_to_rlvl"][-1]}')

    perps_dash = pd.DataFrame(perps_dash)

    code_completed()
    return perps_dash
```

```python
dash = get_perp_dash()
code_completed()
```
